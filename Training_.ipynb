## CELL 5 — Fix the inference file (torch.load bug)

This resets the inference file completely and sets `weights_only=False` exactly once.

```python
import re
from pathlib import Path

inf_file = Path('/content/NuLite/nuclei_detection/inference/inference_nufastvit_experiment_pannuke.py')
text = inf_file.read_text()

# Step 1: remove ALL existing weights_only arguments entirely (however many there are)
text = re.sub(r',\s*weights_only\s*=\s*\w+', '', text)

# Step 2: now add weights_only=False exactly once to every torch.load call
text = re.sub(
    r'(torch\.load\([^)]+)(\))',
    r'\1, weights_only=False\2',
    text
)

inf_file.write_text(text)

# Verify
print('torch.load lines after fix:')
for i, line in enumerate(text.splitlines()):
    if 'torch.load' in line:
        print(f'  line {i+1}: {line.strip()}')
print('Done.')
```

## CELL 6 — Fix the experiment file (hardcoded dataloaders + paths)

```python
import re
from pathlib import Path

exp_file = Path('/content/NuLite/nuclei_detection/experiments/experiment_nulite_pannuke.py')
text = exp_file.read_text()

# Fix ALL num_workers and pin_memory in this file
text = re.sub(r'num_workers\s*=\s*\d+', 'num_workers=0', text)
text = re.sub(r'pin_memory\s*=\s*True', 'pin_memory=False', text)

# Fix hardcoded val batch_size=128 -> 4
text = re.sub(r'batch_size\s*=\s*128', 'batch_size=4', text)

exp_file.write_text(text)

print('Experiment file fixed. DataLoader lines:')
for i, line in enumerate(text.splitlines()):
    if any(k in line for k in ['num_workers', 'pin_memory', 'batch_size=4', 'batch_size=128']):
        print(f'  line {i+1}: {line.strip()}')
```

## CELL 7 — Write the training config YAML (2 epochs, batch size 4)

```python
import yaml, os

os.makedirs('/content/NuLite/logs', exist_ok=True)

config = {
    'data': {
        'dataset': 'PanNuke',
        'dataset_path': '/content/NuLite/data/pannuke/data_processed',
        'num_nuclei_classes': 6,
        'num_tissue_classes': 19,
        'train_folds': [0],
        'val_folds':   [1],
        'test_folds':  [2],
    },
    'eval_checkpoint': 'model_best.pth',
    'gpu': 0,
    'logging': {
        'group':      'NuLite-fastvit_t8',
        'level':      'Debug',
        'log_comment':'NuLite-fastvit_t8-Fold-0',
        'log_dir':    '/content/NuLite/logs',
        'mode':       'disabled',
        'notes':      'NuLite',
        'project':    'NuLite',
        'tags':       ['Fold-0', 'fastvit_t8'],
        'wandb_dir':  '/content/NuLite/logs',
    },
    'loss': {
        'hv_map': {
            'mse':  {'loss_fn': 'mse_loss_maps',  'weight': 1},
            'msge': {'loss_fn': 'msge_loss_maps', 'weight': 1},
        },
        'nuclei_binary_map': {
            'dice':            {'loss_fn': 'dice_loss',        'weight': 1},
            'focaltverskyloss':{'loss_fn': 'FocalTverskyLoss', 'weight': 1},
        },
        'nuclei_type_map': {
            'bce':               {'loss_fn': 'xentropy_loss',      'weight': 1},
            'dice':              {'loss_fn': 'dice_loss',           'weight': 1},
            'mcfocaltverskyloss':{'loss_fn': 'MCFocalTverskyLoss',  'weight': 1, 'args': {'num_classes': 6}},
        },
        'tissue_types': {
            'ce': {'loss_fn': 'CrossEntropyLoss', 'weight': 1},
        },
    },
    'model': {'backbone': 'fastvit_t8'},
    'random_seed': 0,
    'run_sweep': False,
    'training': {
        'batch_size':               4,      # <-- batch size 4
        'drop_rate':                0,
        'early_stopping_patience':  2,
        'epochs':                   2,      # <-- 2 epochs
        'mixed_precision':          True,
        'optimizer':                'AdamW',
        'optimizer_hyperparameter': {
            'betas':        [0.85, 0.95],
            'lr':           0.0003,
            'weight_decay': 0.0001,
        },
        'sampling_gamma':    0.85,
        'sampling_strategy': 'cell+tissue',
        'scheduler': {
            'hyperparameters': {'gamma': 0.85},
            'scheduler_type':  'exponential',
        },
        'unfreeze_epoch': 0,
    },
    'transformations': {
        'blur':             {'blur_limit': 11, 'p': 0.2},
        'colorjitter':      {'p': 0.2, 'scale_color': 0.1, 'scale_setting': 0.25},
        'downscale':        {'p': 0.15, 'scale': 0.5},
        'elastictransform': {'p': 0.2},
        'gaussnoise':       {'p': 0.25, 'var_limit': 50},
        'horizontalflip':   {'p': 0.5},
        'normalize':        {'mean': [0.5, 0.5, 0.5], 'std': [0.5, 0.5, 0.5]},
        'randomrotate90':   {'p': 0.5},
        'randomsizedcrop':  {'p': 0.0},  # disabled to avoid float/int bug
        'superpixels':      {'p': 0.1},
        'verticalflip':     {'p': 0.5},
        'zoomblur':         {'p': 0.1},
    },
}

with open('/content/NuLite/config_colab.yaml', 'w') as f:
    yaml.dump(config, f, default_flow_style=False)

print('Config written. Key settings:')
print(f'  epochs     = {config["training"]["epochs"]}')
print(f'  batch_size = {config["training"]["batch_size"]}')
print(f'  dataset    = {config["data"]["dataset_path"]}')
print(f'  wandb      = {config["logging"]["mode"]}')
```
